```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact
```

```python
f1=10
f2=50
A1=1
A2=1



t=np.linspace(0,1,1025)
t_c=np.linspace(0,1,1025*2-1)
s= A1*np.sin(np.pi*2*f1*t) + A2*np.sin(np.pi*2*f2*t)



```

```python
def Convol_plot(freq):
  s_var= np.sin(np.pi*2*freq*t)
  Cv_= np.convolve(s,s_var)

  # Crear la figura y los subgráficos
  fig, axs = plt.subplots(2, 2, figsize=(10, 6))

  # Graficar en la primera fila (dos gráficos)
  axs[0, 0].plot(t, s, color='b')
  axs[0, 0].set_title("Signal 1:combination")

  axs[0, 1].plot(t, s_var, color='r')
  axs[0, 1].set_title("Signal 2:variable signal")
  # Unir las dos columnas de la segunda fila en una sola
  ax_big =fig.add_subplot(2, 1, 2)
  ax_big.plot(t_c, Cv_, color='g')
  ax_big.set_title("S * Svar")
  ax_big.set_xlim(0, 1)  # Rango del eje X
  ax_big.set_ylim(-500, 500)  # Rango del eje X

  # Eliminar los subgráficos innecesarios de la segunda fila
  fig.delaxes(axs[1, 0])
  fig.delaxes(axs[1, 1])


  # Ajustar el diseño
  plt.tight_layout()
  plt.show()
```

```python
interact(Convol_plot, freq=widgets.FloatSlider(min=0.0, max=100, step=1, value=1, description="Frecuencia"))
```

##Modificación función para hacer y mostrar la convolución de mi señal original y una función enventanada, a través de un gráfico interactivo, para simulación de wavelets

```python
def signal_window(s,li,with__,t):
  w = np.zeros_like(t)
  for i in range(0,len(w)):
    if(t[i] > li  and t[i]<li+with__):

        w[i]=1
    else:
      w[i]=0

  return s*w
```

```python
def Convol_plot_limit(freq, with_,li):

  s_var= np.sin(np.pi*2*freq*t)
  svar=signal_window(s_var,li,with_,t)
  Cv_= np.convolve(s,svar)

  # Crear la figura y los subgráficos
  fig, axs = plt.subplots(2, 2, figsize=(10, 6))

  # Graficar en la primera fila (dos gráficos)
  axs[0, 0].plot(t, s, color='b')
  axs[0, 0].set_title("Signal 1:combination")

  axs[0, 1].plot(t, svar, color='y')
  axs[0, 1].set_title("Signal 2:variable signal")
  # Unir las dos columnas de la segunda fila en una sola
  ax_big =fig.add_subplot(2, 1, 2)
  ax_big.plot(t_c, Cv_, color='g')
  ax_big.set_title("S * Svar")
  ax_big.set_xlim(0, 1)  # Rango del eje X
  ax_big.set_ylim(-500, 500)  # Rango del eje X

  # Eliminar los subgráficos innecesarios de la segunda fila
  fig.delaxes(axs[1, 0])
  fig.delaxes(axs[1, 1])


  # Ajustar el diseño
  plt.tight_layout()
  plt.show()
```

```python
#usando una señal limitada en el tiempo


interact(Convol_plot_limit, freq=widgets.FloatSlider(min=0.0, max=100, step=1, value=1, description="Frecuencia"),
         with_=widgets.FloatSlider(min=0, max=1, step=.1, value=1, description="Ancho de onda"),
         li=widgets.FloatSlider(min=0, max=1, step=.1, value=0, description="limite inferior de onda"))
```
