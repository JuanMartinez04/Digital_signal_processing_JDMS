#PSD cambiando el número de muestras

```python
#importamos librerías a utilizar
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

```python
#Definimos nuestras señales
A=1
frec= 10
fs_1=20 #frecuencia de muestreo 1
fs_2= 100 #frecuencia de muestreo 2
t_i=0 #tiempo inicial
t_f=1 #tiempo final
t_1= np.linspace(t_i,t_f,fs_1*t_f)
t_2= np.linspace(t_i,t_f,fs_2*t_f)

f_1= A*np.sin(2*np.pi*frec*t_1)
f_2= A*np.sin(2*np.pi*frec*t_2)

#dibujamos las señales
plt.figure(figsize=(8,4))
plt.plot(t_1,f_1,'red')
plt.plot(t_2,f_2)
```

la imagen anterior se distorciona al tener tan pocos datos

#Calculamos la fft de las señales

```python
#calculamos la fft
fft_1= sc.fft.fft(f_1)
fft_2=sc.fft.fft(f_2)
#dibujamos el espectro resultante sin vector de frecuencias
plt.figure(figsize=(8,8))
plt.title('PSD de las señales(sin vector de frecuencias)')
plt.plot(np.abs(fft_1),'red',label='PSD señal de pocas muestras')
plt.plot(np.abs(fft_2),label='PSD señal de suficientes muestras')
plt.legend(loc='upper right')
plt.grid()
```

```python
#sacamos el vector de frecuencias
frecs_1= sc.fft.fftfreq(len(fft_1),1/fs_1)
frecs_2= sc.fft.fftfreq(len(fft_2),1/fs_2)
#deibujamos el espectro con el vector de frecuencias y con la magnitud de la potencia
plt.figure(figsize=(8,8))
plt.title('PSD de las señales')
plt.plot(frecs_1,np.abs(fft_1),'red',label='PSD de la señal de pocas muestras')
plt.plot(frecs_2,np.abs(fft_2),label='PSD de la señal con más muestras')
plt.legend(loc='upper right')
plt.grid()
```
